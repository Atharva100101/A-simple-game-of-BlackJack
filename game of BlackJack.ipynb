Class Definations

Consoder making a Card class where each Card object has a suit or rank, then a deck class to hold all the 52 Card object, and can be shuffled, and finally a Hand class which can hold those Cards that have been dealt to each player from the Deck. 


```python
''' Step2: Create a Card Class 
A card object really only needs two attributes: suits and ranks. You might add an attribute for VALUE- we would handling the
the value situation later when we will make Hand class. In additon to Card__init__method, consider adding a __str__ method 
that, when asked to print a card return a string something like- Two of hearts or three of diamonds'''

class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
    def __str__(self):
        return self.rank + ' of ' +  self.suit
```

```python
'''Step 3: Create a deck class
Here we might store 52 unique card object in a list that can be shuffled with the help of importing random module and then 
using the function random.shuffle(). In order to do that first we need to instantiate 52 unique card object and add them in 
our list. So long as the Card class definition appears in our code, we can build Card objects inside our Deck __init__ method.
Sample:
for suit in suits:
    for rank in ranks'''

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))  # build Card objects and add them to the list
    
    def __str__(self):
        deck_comp = ''  # start with an empty string
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # add each Card object's print string
        return 'The deck has:' + deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
        
```

```python
'''Yup! Now lets tackle the issue of ace adjustment. If a hand exceeds a value 21 but it contian an Ace, we can reduce the 
value from 11 to 1 and continue playing'''
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank=='Ace':
            self.aces+=1
            
    def adjust_for_ace(self):
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
```

```python
'''step 5 creating a chip class 
In addition to Deck and Card class that we have made in this code and the previous sample code, we also need to keep a track 
on players chips, bet and ongoing wins. This could be done through global variable, but in the spirit of OOPS and to understand 
the concept clearly lets make the chips class instead'''



class Chips:
    def __init__(self):
        self.total=100
        self.bet=0
        
    def win_bet(self):
        self.total+=self.bet
        
    def lose_bet(self):
        self.total-=self.bet
        
'''A note on provding default value
Alternatively, we could have passed a default total value as an parameter in the __init__. This would have let us pass in 
an override value at the time the object was created rather than wait until later to change it. The code would have looked 
like this:

def __init__(self,total=100):
        self.total = total
        self.bet = 0

Either technique is fine, it only depends on how you plan to start your game parameters.'''

```

```python
'''step 6 '''
'''Function defination
A lot od steps are going to be repetetive like taking bet ans showing the cards that the player and the computer has. In order
to accomodate these repetitive task we use the functions. The follwing steps are guideline. Fell free to add or remove the 
function based on your preferences'''

def take_bet(chips):
    while True:
        try:
            chips.bet=int(input('Enter the number of chips you want to bet'))
        except ValueError:
            print('There can only be an integer value')
        else:
            if chips.bet>chips.total:
                print('Sorry! Your bet cannot exceed', chips.total)
            else:
                break
```

```python
'''step 7 
Write a function for taking hit
Either player can take hits until they bust. This function will be called during gameplay anytime a Player requests a hit, 
or a Dealer's hand is less than 17. It should take in Deck and Hand objects as arguments, and deal one card off the deck 
and add it to the Hand. You may want it to check for aces in the event that a player's hand exceeds 21.'''

def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
'''step 8: write a functin which would prompt a player to hit or stand
This function should accept the deck and the player's hand as arguments, and assign playing as a global variable.
If the Player Hits, employ the hit() function above. If the Player Stands, set the playing variable to False - this will 
control the behavior of a while loop later on in our code.'''


def hit_or_stand(deck,hand):
    global playing
    
    while True:
        x=input('Do you want to hit or stand: h or s ')
        if x[0].lower()=='h':
            return hit(deck,hand)
        elif x[0].lower()=='s':
            print('Player stand. Dealer is playing')
            playing=False
        else:
            print('Sorry please try agin later')
            continue
        break
```

```python
'''step 9: Write a function to display the cards
When the game starts, and after each time Player takes a card, the dealer's first card is hidden and all of Player's cards are visible. At the end of the hand all cards are shown, and you may want to show each hand's total value.
Write a function for each of these scenarios.'''

def show_some(player,dealer):
    print("\n Dealer's Hand: ")
    print("<card hidden>")
    print('',dealer.cards[1])
    print("\n Player's Hand:", *player.cards, sep='\n ')
    
def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("Player's Hand =",player.value)
```
